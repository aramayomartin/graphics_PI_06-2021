```python
#En el plano xy
aux_x=[] #Lista para graficar despues 
aux_y=[]
#aux_z=[]
aux_atr=[]
#---------------------------------------------------------#
#PARA SELECCIONAR LAS DIMENSIONES DE LA PORQUERIA
for i in range(len(datos.x)):
    dist = np.sqrt(datos.x[i]**2+datos.y[i]**2) #Esto es para achicar el corte a solo la zona de opera
    if datos.z[i] == 163.3: #and dist > 415 and dist <515:
        aux_x.append(datos.x[i])
        aux_y.append(datos.y[i])
        #aux_z.append(datos.z[i])
        aux_atr.append(datos.Dosis[i])
#------------------------------------------------------------------
#PARA GRAFICAR - AUXILIARES - NO TOCAR
Graf = {'x':aux_x,'y':aux_y,'Dosis':aux_atr}
df=pd.DataFrame(Graf)

Dosis = df.pivot_table(index='x', columns='y', values='Dosis').T.values
X_unique = np.sort(df.x.unique())
Y_unique = np.sort(df.y.unique())
X, Y = np.meshgrid(X_unique, Y_unique)
#------------------------------------------------------------------
#PARA TOCAR EL GRAFICO
fig, ax = plt.subplots()
s = ax.pcolor(X,Y,Dosis, shading='auto')
fig.colorbar(s, ax=ax)
ax.set_ylabel('y $[cm]$')
ax.set_xlabel('x $[cm]$')
#plt.xlim(20,60)
ax.set_title('Tasa de Dosis $[\mu Gy/h]$')   
```
